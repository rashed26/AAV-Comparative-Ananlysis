#Final Supervised Learning

```python
import os
import numpy as np
import torch
import torchvision.transforms as transforms
from torchvision import models
from torch.utils.data import DataLoader, Dataset
from PIL import Image
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, auc, classification_report, accuracy_score, confusion_matrix
from sklearn.preprocessing import label_binarize
import seaborn as sns
import matplotlib.pyplot as plt

# Define dataset class
class ImageDataset(Dataset):
    def __init__(self, image_folder, transform=None):
        self.image_folder = image_folder
        self.transform = transform
        self.image_files = []
        self.labels = []

        for label, class_name in enumerate(os.listdir(image_folder)):
            class_folder = os.path.join(image_folder, class_name)
            if os.path.isdir(class_folder):
                for img_name in os.listdir(class_folder):
                    img_path = os.path.join(class_folder, img_name)
                    self.image_files.append(img_path)
                    self.labels.append(label)

    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        img_path = self.image_files[idx]
        image = Image.open(img_path).convert('RGB')
        label = self.labels[idx]

        if self.transform:
            image = self.transform(image)

        return image, label

# Image transformations
transform = transforms.Compose([
    transforms.Resize((299, 299)),  # Updated for InceptionV3 compatibility
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Load image dataset
image_folder = '/content/drive/MyDrive/Sahadat Sir/train_val_photos'
dataset = ImageDataset(image_folder, transform=transform)
dataloader = DataLoader(dataset, batch_size=32, shuffle=False)

# Load CNN model

def load_model(model_name):
    if model_name == 'resnet50':
        model = models.resnet50(pretrained=True)
        model = torch.nn.Sequential(*list(model.children())[:-1])
    elif model_name == 'vgg16':
        model = models.vgg16(pretrained=True)
        model = torch.nn.Sequential(*list(model.children())[:-1])
    elif model_name == 'resnet18':
        model = models.resnet18(pretrained=True)
        model = torch.nn.Sequential(*list(model.children())[:-1])
    elif model_name == 'mobilenet_v2':
        model = models.mobilenet_v2(pretrained=True)
        model = torch.nn.Sequential(*list(model.features))
    elif model_name == 'inception_v3':
        from torchvision.models import Inception_V3_Weights
        base_model = models.inception_v3(weights=Inception_V3_Weights.DEFAULT, aux_logits=True)
        base_model.fc = torch.nn.Identity()  # Remove final classification layer
        model = base_model
    else:
        raise ValueError("Unsupported model type")
    model.eval()
    return model

# Extract features

def extract_features(dataloader, model):
    features = []
    targets = []
    with torch.no_grad():
        for images, labels in dataloader:
            outputs = model(images)
            features.append(outputs.view(outputs.size(0), -1).numpy())
            targets.extend(labels.numpy())
    return np.vstack(features), np.array(targets)

# Classifier selector
def get_classifier(name):
    if name == 'svm': return SVC(kernel='linear', probability=True)
    elif name == 'random_forest': return RandomForestClassifier(n_estimators=100, random_state=42)
    elif name == 'knn': return KNeighborsClassifier(n_neighbors=3)
    else: raise ValueError("Unknown classifier")

# Evaluation routine
def evaluate_model(X, y, classifier_name, model_name):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    clf = get_classifier(classifier_name)
    for epoch in range(5):
        print(f"Epoch {epoch+1}/5")
        clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    y_score = clf.predict_proba(X_test)

    print(f"\n=== {model_name.upper()} + {classifier_name.upper()} ===")
    print(classification_report(y_test, y_pred))
    print("Accuracy:", accuracy_score(y_test, y_pred))

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(f"Confusion Matrix - {model_name.upper()} + {classifier_name.upper()}")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()

    # ROC Curve
    y_test_bin = label_binarize(y_test, classes=np.unique(y))
    plt.figure(figsize=(6, 6))
    for i in range(y_score.shape[1]):
        fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_score[:, i])
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f"Class {i} (AUC={roc_auc:.2f})")
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"ROC Curve - {model_name.upper()} + {classifier_name.upper()}")
    plt.legend(loc='lower right')
    plt.grid()
    plt.show()

# Run for all models and only SVM classifier
feature_extractors = ['resnet50', 'vgg16', 'resnet18', 'mobilenet_v2', 'inception_v3']
classifiers = ['svm', 'random_forest', 'knn']

print("DEBUG - feature_extractors:", feature_extractors)
print("DEBUG - type:", type(feature_extractors))

for model_name in feature_extractors:
    try:
        model = load_model(model_name)
        X, y = extract_features(dataloader, model)
        for classifier_name in classifiers:
            evaluate_model(X, y, classifier_name, model_name)
    except Exception as e:
        print(f"Error with {model_name}: {e}")
    except Exception as e:
        print(f"Error with {model_name}: {e}")
    except Exception as e:
        print(f"Error with {model_name}: {e}")
```
